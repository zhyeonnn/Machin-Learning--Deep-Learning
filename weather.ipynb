```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as seabornInstance
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
dataset = pd.read_csv(r'c:\Users\인공지능그랜드ICT\Documents\weather.csv')
dataset.plot(x='MinTemp', y='MaxTemp',style='o')
plt.title('MinTemp vs MaxTemp')
plt.xlabel('MinTemp')
plt.ylabel('MaxTemp')
plt.show()
```

```python
# reshape(-1,1)은 행 -> 자동(값이 있는 만큼) , 열 -> 1줄로.
X=dataset['MinTemp'].values.reshape(-1,1)
y=dataset['MaxTemp'].values.reshape(-1,1)
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)

regressor = LinearRegression()
regressor.fit(X_train, y_train)
# X_test값에 대한 예측값
y_pred = regressor.predict(X_test)
print("y_pred -> \n",y_pred)

# 데이터프레임으로 묶음(실제값, 예측값)
df = pd.DataFrame({'Actual':y_test.flatten(),'Predicted':y_pred.flatten()})
print("df -> \n",df)
```

```python
# 회귀선으로 표현(직선)
# 실제 데이터 값
plt.scatter(X_test, y_test,color='gray')
# 예측한 데이터 값
plt.plot(X_test,y_pred,color='red',linewidth =2)
plt.show()
# 예측값이 실제 데이터 값과 비슷한 걸 알 수 있음
# 실제값과 예측값에 대한
# mean_squared_error:평균 제곱 오차
print('평균 제곱법: ', metrics.mean_squared_error(y_test,y_pred))
# np.sqrt(mean_squared_error()):루트 평균 제곱 오차
print('루트 평균제곱법: ',np.sqrt(metrics.mean_squared_error(y_test,y_pred)))

# 모델의 정확도를 알 수 있는 방법
# 전체 데이터의 5~10%이하면 예측률이 좋은 것
maxTemp = dataset['MaxTemp'].values.max()
minTemp = dataset['MinTemp'].values.min()
tempRange = maxTemp - minTemp
succesePedict = tempRange * 0.1
print("10% 기준값 :",succesePedict)
```
